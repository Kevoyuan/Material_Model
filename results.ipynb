```python
import os

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats
```

Arm Degree

```python
Arm_Angle = "./arm_Angle/test.csv"
df_Arm_Angle = pd.read_csv(Arm_Angle)

x_ArmAngle = df_Arm_Angle["ArmAngle"]
y_ArmAngle = df_Arm_Angle["Eq_Strains2"]
y_state_ArmAngle = df_Arm_Angle["State"]

plt.xticks(np.linspace(50, 60, 11))

res_ArmAngle = stats.linregress(x_ArmAngle, y_ArmAngle)
print(f"R-squared: {res_ArmAngle.rvalue**2:.6f}")


plt.plot(x_ArmAngle, y_ArmAngle, "o", label="original data")
plt.plot(
    x_ArmAngle,
    res_ArmAngle.intercept + res_ArmAngle.slope * x_ArmAngle,
    "r",
    label="fitted line",
)
plt.legend()
plt.ylim([0.02, 0.08])
plt.xlabel("Arm Angle/[°]")
plt.ylabel("Eq_Strains/[MPa]")

ymax_ArmAngle = max(y_ArmAngle)
xpos_ArmAngle = y_ArmAngle.idxmax()
# print(xpos_ArmAngle)
xmax_ArmAngle = x_ArmAngle.iloc[xpos_ArmAngle]

plt.annotate(
    f"max={xmax_ArmAngle,round(ymax_ArmAngle,5)}",
    xy=(xmax_ArmAngle, ymax_ArmAngle),
    xytext=(xmax_ArmAngle, ymax_ArmAngle + 0.005),
    arrowprops=dict(
        arrowstyle="->", facecolor="black", connectionstyle="arc3,rad=-0.2"
    ),
    horizontalalignment="right",
    verticalalignment="bottom",
)

plt.annotate(
    f"R\u00b2={res_ArmAngle.rvalue**2:.6f}",
    xy=(51, 0.077),
    va="center",
    ha="center",
    bbox=dict(boxstyle="round", fc="w"),
)

for i, txt in enumerate(y_state_ArmAngle):
    plt.annotate(txt, (x_ArmAngle[i], y_ArmAngle[i]))

plt.savefig("./arm_Angle/Arm_Angle.png")
a = plt.show()
```
